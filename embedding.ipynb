```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### Sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
sent
```

```python
### Define the vocab size
voc_size = 10000
```

```python
## One hot representation of every word
one_hot_repr = [one_hot(words,voc_size)for words in sent]
one_hot_repr

#-> this will give index of each word where it is present in 10000 voc
```

```python
## Word Embedding representation
from tensorflow.keras.layers import Embedding,Input
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## We need to make all sent of same size -> we use pad_sequences and set max sent_len add 
## zeroes if sent length less than max-> it is called prepadding/postpadding (zeroes kha ayegi age ya picche)
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen = sent_length)
print(embedded_docs)
```

```python
## Feature Representation
dim=10
```

```python
model = Sequential([
    Input(shape=(sent_length,)),
    Embedding(input_dim=voc_size,output_dim=dim)
])
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
